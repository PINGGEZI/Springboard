```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
engine = create_engine('sqlite:///sqlite_db_pythonsqlite.db')
```

**Q10: Produce a list of facilities with a total revenue less than 1000.
The output of facility name and total revenue, sorted by revenue. Remember
that there's a different cost for guests and members!**

```python
query = 'SELECT f.name AS Facility,\
                sum(CASE WHEN b.memid =0\
                    THEN b.slots * f.guestcost\
                    ELSE b.slots * f.membercost\
                    END) AS revenue\
         FROM Bookings AS b\
         INNER JOIN Facilities AS f\
         USING (facid)\
         GROUP BY f.name\
         Having revenue <1000\
         ORDER BY revenue DESC'
```

```python
pd.read_sql_query(query,engine)
```

**Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order**


```python
query = 'SELECT a.surname AS surname,\
                a.firstname AS firstname,\
                b.firstname ||" " || b.surname AS recommendedby\
         FROM Members AS a\
         LEFT JOIN Members AS b\
         ON a.recommendedby = b.memid\
         WHERE a.memid != 0\
         ORDER BY a.surname,a.firstname'
```

```python
pd.read_sql_query(query,engine)
```

**Q12: Find the facilities with their usage by member, but not guests**


```python
query = 'SELECT m.firstname || " " || m.surname AS member,\
                f.name AS facilities,\
                total_slots\
         FROM(\
             SELECT memid,\
             facid,\
             SUM(slots) AS total_slots\
             FROM Bookings\
             WHERE memid != 0\
             GROUP BY memid,facid\
             ) AS u\
         LEFT JOIN Facilities AS f\
         USING(facid)\
         LEFT JOIN Members AS m\
         USING(memid)\
         ORDER BY member DESC'
```

```python
pd.read_sql_query(query,engine)
```

**Q13: Find the facilities usage by month, but not guests**


```python
query = 'SELECT STRFTIME("%m",b.starttime) AS month,\
                f.name AS facility,\
                SUM(slots) AS total_slots\
        FROM Bookings AS b\
        LEFT JOIN Facilities AS f\
        USING(facid)\
        WHERE b.memid != 0\
        GROUP BY STRFTIME("%m",b.starttime),f.name'
```

```python
pd.read_sql_query(query,engine)
```
