**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(dfg):
    p_1998 = float(dfg.Avg_Price[dfg.Year == 1998])
    p_2018 = float(dfg.Avg_Price[dfg.Year == 2018])
    return p_2018/p_1998

```

```python
price_ratio = {}

#iterate through all the unique London_Boroughs 
#work out the ratio capturing the difference of house prices between 1998 and 2018

for b in price_byear.Boroughs.unique():
    dfg = price_byear[price_byear['Boroughs'] == b]
    price_ratio[b] = create_price_ratio(dfg)
    
    
price_ratio_df  = pd.DataFrame.from_dict(price_ratio, orient='index')
price_ratio_df.reset_index(inplace = True)
```

```python
price_ratio_df.rename(columns = {'index' : 'Boroughs',0 : 'Price_Ratio'},inplace = True)
```

```python
price_ratio_df.head()
```

```python
# find out which the boroughs in london has the greatest increase in housing prices, on average, over the last two decades
#sort the price ratio in descending order
sorted_price = price_ratio_df.sort_values(by = 'Price_Ratio',ascending = False)

sorted_price

```

```python
#plot the top 10 boroughs that have the greatest increase in housing prices
fig = sorted_price[:15][::-1].plot(kind = 'barh',x= 'Boroughs',y = 'Price_Ratio',figsize=(10,6),sort_columns=True)
fig.set_title('Top 10 Boroughs with the greatest housing prices increase')
```

```python
#plot the TOP 10 boroughs that have the least increase in housing prices,which help us 
#understand the underlying factors affect the housing prices. 
fig = sorted_price[22:].plot(kind = 'barh',x= 'Boroughs',y = 'Price_Ratio',figsize=(10,6),sort_columns=True)
fig.set_title('TOP 10 Boroughs with the Least housing prices increase')
```
